```python
#zipfile_path = "lang-8/lang-8.zip"
#zf_files = zipfile.ZipFile(zipfile_path, "r")# zipfile_path = "lang-8/lang-8.zip"
native_language_list = [] 
raw_text_list = []
if filename[-4:] == "html":
    filename = open(filename, "r", encoding="utf-8")
    soup = BeautifulSoup(filename, 'html.parser')
    for li in soup.find_all('li', {"class":'speaking', "title":True}):
        native_language_list.append(li.contents)
    for text in soup.find_all("div", {"id": "body_show_ori"}):
        raw_text_list.append(text.get_text())
        file_info = {"Filename":filename.name, "Native Language": native_language_list, "Raw Text": raw_text_list}
    pd.DataFrame(file_info)
```
